```python
import random

def adivina_el_numero():
    """
    FunciÃ³n principal para el juego 'Adivina el NÃºmero'.
    """

    # --- 1. ConfiguraciÃ³n del juego ---
    print("==============================================")
    print("Â¡Bienvenido al juego 'Adivina el NÃºmero'! ğŸ²")
    print("==============================================")
    print("\nHe pensado en un nÃºmero entre 1 y 100.")
    print("Â¿PodrÃ¡s adivinar cuÃ¡l es? Tienes 10 intentos.")

    # La computadora elige un nÃºmero secreto al azar
    numero_secreto = random.randint(1, 100)
    intentos_realizados = 0
    intentos_maximos = 10

    # --- 2. Bucle principal del juego ---
    while intentos_realizados < intentos_maximos:
        print(f"\n--- Intento #{intentos_realizados + 1} ---")

        # Se solicita al jugador que ingrese un nÃºmero
        try:
            intento_jugador = int(input("Ingresa tu nÃºmero: "))
        except ValueError:
            # Manejo de error si el jugador no ingresa un nÃºmero
            print("âŒ Â¡Error! Debes ingresar un nÃºmero vÃ¡lido.")
            continue # Vuelve al inicio del bucle

        # --- 3. ComprobaciÃ³n de la respuesta ---
        if intento_jugador < numero_secreto:
            print("ğŸ“ˆ El nÃºmero que pensÃ© es mÃ¡s ALTO.")
        elif intento_jugador > numero_secreto:
            print("ğŸ“‰ El nÃºmero que pensÃ© es mÃ¡s BAJO.")
        else:
            # Si el jugador adivina
            print(f"\nğŸ‰ Â¡Felicidades! Â¡Adivinaste el nÃºmero en {intentos_realizados + 1} intentos!")
            print(f"El nÃºmero secreto era {numero_secreto}.")
            break # Termina el bucle porque el jugador ganÃ³

        intentos_realizados += 1

    # --- 4. Fin del juego ---
    # Este cÃ³digo se ejecuta si el bucle 'while' termina sin que el jugador adivine
    if intentos_realizados == intentos_maximos:
        print("\nğŸ’¥ Â¡Oh no! Te quedaste sin intentos.")
        print(f"El nÃºmero secreto era {numero_secreto}. Â¡Mejor suerte la prÃ³xima vez!")

# --- Ejecutar el juego ---
adivina_el_numero()
```
