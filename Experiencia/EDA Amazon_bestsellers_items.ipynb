### 7. Resumen y hallazgos

Aquí presentamos un resumen de los hallazgos clave de nuestro Análisis Exploratorio de Datos:

*   **Limpieza de datos:** Se eliminaron filas con valores faltantes en 'product_price', 'product_star_rating' y 'product_num_ratings', y se verificó la ausencia de duplicados. La columna 'rank_change_label' se eliminó por completo debido a la falta de datos.
*   **Análisis Univariado:**
    *   La distribución de precios (`product_price_numeric`) muestra una alta concentración en precios bajos, con algunos valores atípicos en rangos de precios más altos.
    *   La mayoría de los productos tienen calificaciones de estrellas entre 3.5 y 4.5, con una distribución sesgada hacia calificaciones más altas.
    *   El número de calificaciones (`product_num_ratings`) está muy sesgado hacia valores bajos, lo que indica que la mayoría de los productos tienen pocas calificaciones, mientras que unos pocos productos tienen un número muy elevado de calificaciones.
    *   La distribución de productos por país muestra una cobertura relativamente uniforme, con India (IN), Alemania (DE), España (ES), Canadá (CA), Estados Unidos (US), Francia (FR) e Italia (IT) teniendo la mayor cantidad de productos en el conjunto de datos.
*   **Análisis Bivariado y Multivariado:**
    *   No se observa una correlación lineal fuerte entre el precio del producto y la calificación de estrellas.
    *   Existe una tendencia a que los productos con un mayor número de calificaciones tengan calificaciones de estrellas más altas, aunque hay excepciones.
    *   El precio promedio varía significativamente entre países, siendo Japón (JP) el que tiene el precio promedio más alto, mientras que Estados Unidos (US) y Canadá (CA) tienen los precios promedio más bajos en este conjunto de datos. Es importante considerar que las monedas y el poder adquisitivo varían entre países, lo que puede influir en esta comparación.
    *   La calificación promedio de estrellas es relativamente similar entre la mayoría de los países, con ligeras variaciones. India (IN) y Alemania (DE) tienen las calificaciones promedio más altas, mientras que España (ES) tiene la calificación promedio más baja.
    *   El número promedio de calificaciones por país también varía, con India (IN), Alemania (DE) y Estados Unidos (US) teniendo un número promedio significativamente mayor de calificaciones por producto en comparación con otros países. Esto podría indicar diferencias en el volumen de ventas o en la cultura de dejar reseñas en cada país.

Estos hallazgos iniciales nos dan una visión general de los datos y las relaciones entre algunas variables clave. Dependiendo de los objetivos específicos del análisis, se podrían realizar investigaciones más profundas en áreas particulares.
