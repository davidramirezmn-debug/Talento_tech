```python
import numpy as np
import random

# Generate 80 additional random data points
np.random.seed(42) # for reproducibility

# List of common Latin American first names
latin_american_names = [
    "Sofia", "Mateo", "Valentina", "Sebastian", "Isabella", "Leonardo", "Camila", "Gabriel", "Valeria", "Nicolas",
    "Daniela", "Alejandro", "Mariana", "Santiago", "Gabriela", "Manuel", "Victoria", "Diego", "Carolina", "Emiliano",
    "Lucia", "Benjamn", "Andrea", "Joaqun", "Paula", "Felipe", "Natalia", "Ignacio", "Alexandra", "Cristobal",
    "Martina", "Agustin", "Fernanda", "Miguel", "Florencia", "Alonso", "Emilia", "Bruno", "Catalina", "Antonio",
    "Josefa", "Agustin", "Isidora", "Vicente", "Antonia", "Maximiliano", "Francisca", "Tomas", "Javiera", "Martin",
    "Jose", "Trinidad", "Gonzalo", "Amanda", "Benjamin", "Paz", "Cristian", "Consuelo", "Esteban", "Constanza",
    "Diego", "Rocio", "Francisco", "Barbara", "Juan", "Monserrat", "Luis", "Jose", "Claudia", "Rodrigo",
    "Alejandra", "Mauricio", "Pilar", "Alfredo", "Bernardita", "Victor", "Daniela", "Patricio", "Ignacia", "Hernan"
]

# Shuffle the list of names
random.shuffle(latin_american_names)

additional_nombres = latin_american_names[:80]
# Generate weights as two-digit numbers
additional_pesos = np.random.randint(30, 99, 80).tolist()
# Generate heights in meters with two decimal places
additional_alturas = np.around(np.random.uniform(1.20, 1.90, 80), decimals=2).tolist()

# Create a new DataFrame with the additional data
additional_df = pd.DataFrame({
    "Nombre": additional_nombres,
    "Peso": additional_pesos,
    "Altura": additional_alturas
})

# Concatenate the original and additional DataFrames
# Reset index to ensure a continuous index after concatenation
df = pd.concat([df.iloc[:20], additional_df], ignore_index=True)


# Calculate BMI (Peso / Altura^2)
df['IMC'] = df['Peso'] / (df['Altura']**2)

# Define a function to classify BMI
def classify_bmi(imc):
    if imc < 18.5:
        return 'Bajo peso'
    elif 18.5 <= imc < 24.9:
        return 'Peso normal'
    elif 24.9 <= imc < 29.9:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

# Apply the classification function to create the 'Clasificacion' column
df['Clasificacion'] = df['IMC'].apply(classify_bmi)

# Display the updated DataFrame
display(df)
```
