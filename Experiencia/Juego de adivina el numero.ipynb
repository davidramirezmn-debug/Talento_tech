```python
import random

def adivina_el_numero():
    """
    Función principal para el juego 'Adivina el Número'.
    """

    # --- 1. Configuración del juego ---
    print("==============================================")
    print("¡Bienvenido al juego 'Adivina el Número'! 🎲")
    print("==============================================")
    print("\nHe pensado en un número entre 1 y 100.")
    print("¿Podrás adivinar cuál es? Tienes 10 intentos.")

    # La computadora elige un número secreto al azar
    numero_secreto = random.randint(1, 100)
    intentos_realizados = 0
    intentos_maximos = 10

    # --- 2. Bucle principal del juego ---
    while intentos_realizados < intentos_maximos:
        print(f"\n--- Intento #{intentos_realizados + 1} ---")

        # Se solicita al jugador que ingrese un número
        try:
            intento_jugador = int(input("Ingresa tu número: "))
        except ValueError:
            # Manejo de error si el jugador no ingresa un número
            print("❌ ¡Error! Debes ingresar un número válido.")
            continue # Vuelve al inicio del bucle

        # --- 3. Comprobación de la respuesta ---
        if intento_jugador < numero_secreto:
            print("📈 El número que pensé es más ALTO.")
        elif intento_jugador > numero_secreto:
            print("📉 El número que pensé es más BAJO.")
        else:
            # Si el jugador adivina
            print(f"\n🎉 ¡Felicidades! ¡Adivinaste el número en {intentos_realizados + 1} intentos!")
            print(f"El número secreto era {numero_secreto}.")
            break # Termina el bucle porque el jugador ganó

        intentos_realizados += 1

    # --- 4. Fin del juego ---
    # Este código se ejecuta si el bucle 'while' termina sin que el jugador adivine
    if intentos_realizados == intentos_maximos:
        print("\n💥 ¡Oh no! Te quedaste sin intentos.")
        print(f"El número secreto era {numero_secreto}. ¡Mejor suerte la próxima vez!")

# --- Ejecutar el juego ---
adivina_el_numero()
```
