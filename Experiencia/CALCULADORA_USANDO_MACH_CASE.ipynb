# CALCULADORA USANDO MACH-CASE



```python
import math


# DEFINICION DE LAS FUNCIONES (Funciones de cálculo sin cambios)

def mostrar_título():

    print('Ejemplo Calculadora Sencilla (+,-,*,/) por David Ramirez:')
    print('='*50)

def mostrar_menu():
    """Muestra el menú de opciones al usuario."""
    print("\nMENU:")
    print("1: Sumar (+)")
    print("2: Restar (-)")
    print("3: Multiplicar (*)")
    print("4: Dividir (/)")
    print("5: Logaritmo (log10)")
    print("6: Coseno (cos)")
    print("7: Tangente (tan)")
    print("8: Seno (sen)")
    print("9: Potenciación (x^y)")
    print("10: Raíz cuadrada")
    print("11: Salir")

# --- Funciones de cálculo ---
def calcular_suma(n1, n2):
    return n1 + n2

def calcular_resta(n1, n2):
    return n1 - n2

def calcular_multiplicacion(n1, n2):
    return n1 * n2

def calcular_division(n1, n2):
    if n2 == 0:
        return "Error: División por cero"
    return n1 / n2

def calcular_logaritmo_base10(n):
    if n <= 0:
        return "Error: Dominio inválido"
    return math.log10(n)

def calcular_coseno(angulo_rad):
    return math.cos(angulo_rad)

def calcular_tangente(angulo_rad):
    return math.tan(angulo_rad)

def calcular_seno(angulo_rad):
    return math.sin(angulo_rad)

def calcular_potenciacion(base, exponente):
    return base ** exponente

def calcular_raiz_cuadrada(n):
    if n < 0:
        return "Error: Número negativo"
    return math.sqrt(n)


def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos):
    """
    Procesa la operación seleccionada por el usuario utilizando match-case
    y muestra el resultado en pantalla.
    """
    # Usamos .lower() para hacer la comparación insensible a mayúsculas
    opcion_normalizada = opcion.lower()

    match opcion_normalizada:
        case "1" | "+" | "sumar":
            print("Vamos a sumar (+) los dos valores ingresados:")
            resultado = calcular_suma(numero_uno, numero_dos)
            print(f"La suma es: {resultado}")

        case "2" | "-" | "restar":
            print("Vamos a restar (-) los dos valores ingresados:")
            resultado = calcular_resta(numero_uno, numero_dos)
            print(f"El resultado es: {resultado}")

        case "3" | "*" | "multiplicar":
            print("Vamos a multiplicar (*) los dos valores ingresados:")
            resultado = calcular_multiplicacion(numero_uno, numero_dos)
            print(f"El resultado es: {resultado}")

        case "4" | "/" | "dividir":
            print("Vamos a dividir (/) los dos valores ingresados:")
            resultado = calcular_division(numero_uno, numero_dos)
            print(f"El resultado es: {resultado}")

        case "5" | "log10" | "logaritmo":
            print("Vamos a calcular el logaritmo (base 10) de los dos valores:")
            print(f"Logaritmo de {numero_uno}: {calcular_logaritmo_base10(numero_uno)}")
            print(f"Logaritmo de {numero_dos}: {calcular_logaritmo_base10(numero_dos)}")

        case "6" | "cos" | "coseno":
            print("Vamos a calcular el coseno (en radianes) de los dos valores:")
            print(f"Coseno de {numero_uno}: {calcular_coseno(numero_uno)}")
            print(f"Coseno de {numero_dos}: {calcular_coseno(numero_dos)}")

        case "7" | "tan" | "tangente":
            print("Vamos a calcular la tangente (en radianes) de los dos valores:")
            print(f"Tangente de {numero_uno}: {calcular_tangente(numero_uno)}")
            print(f"Tangente de {numero_dos}: {calcular_tangente(numero_dos)}")

        case "8" | "sen" | "seno":
            print("Vamos a calcular el seno (en radianes) de los dos valores:")
            print(f"Seno de {numero_uno}: {calcular_seno(numero_uno)}")
            print(f"Seno de {numero_dos}: {calcular_seno(numero_dos)}")

        case "9" | "potenciacion" | "potencia":
            print(f"Vamos a calcular la potencia de {numero_uno} elevado a {numero_dos}:")
            resultado = calcular_potenciacion(numero_uno, numero_dos)
            print(f"El resultado es: {resultado}")

        case "10" | "raiz" | "raiz cuadrada":
            print("Vamos a calcular la raíz cuadrada de los dos valores:")
            print(f"Raíz de {numero_uno}: {calcular_raiz_cuadrada(numero_uno)}")
            print(f"Raíz de {numero_dos}: {calcular_raiz_cuadrada(numero_dos)}")

        # El caso "_" actúa como el "else": se ejecuta si no coincide ninguna opción anterior
        case _:
            print("Opción no válida. Por favor, intente de nuevo.")


#======================================================================
# EJECUCIÓN PRINCIPAL DEL PROGRAMA
#======================================================================
def main():
    """
    Función principal que ejecuta el bucle de la calculadora.
    """
    mostrar_título()

    while True:
        mostrar_menu()
        opcion = input("Ingrese la opción deseada del menú: ")

        # Condición de salida del bucle
        if opcion == "11" or opcion.lower() == "salir":
            print("\n¡Gracias por usar la calculadora!")
            break

        try:
            # Pedir los números al usuario
            numero_uno = float(input("Ingrese el primer número: "))
            numero_dos = float(input("Ingrese el segundo número: "))
            print("-" * 20) # Separador

            # Llamar a la función que ahora usa match-case
            procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos)

        except ValueError:
            print("\n¡Error! Debe ingresar valores numéricos válidos.")
        except Exception as e:
            print(f"\nHa ocurrido un error inesperado: {e}")

        print("\n" + "="*50) # Separador para la siguiente operación

# --- Iniciar el programa ---
main()
```
